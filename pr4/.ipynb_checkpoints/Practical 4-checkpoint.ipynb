```python
plt.figure(figsize=(10,3))
plt.suptitle('Knn Algorithm')
plt.subplot(1,2,1)
plt.title('No. of Neighbors vs Accuracy')
plt.plot(n_neighbors,accs,c='g')
plt.subplot(1,2,2)
plt.title('No. of Neighbors vs Time')
plt.plot(n_neighbors,times,c='r')
plt.show()
```
