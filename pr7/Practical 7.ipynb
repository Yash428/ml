```python
import pandas as pd
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./iris_two.csv',sep=',')
```

```python
df.columns = ['sepal_length','sepal_width','petal_length','petal_width','species']
```

```python
df.head()
```

```python
for i in df['species'].unique():
    print('Total rows of', i,':',len(df[df['species']==i]))
```

```python
df.isnull().sum()
```

```python
sns.pairplot(df, hue='species')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop('species',axis=1)
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
er = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 42)
    kmeans.fit(X)
    er.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), er)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('er')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(X)
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'purple', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'orange', label = 'Iris-versicolour')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'red', marker='X', label = 'Centroids')

plt.legend()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.5,min_samples=5)
y_pred = dbscan.fit_predict(X)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=dbscan.labels_, cmap='viridis', marker='o', edgecolor='k', s=50, alpha=0.7)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('DBSCAN Clustering Results (eps=0.5, min_samples=5)')
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
def evaluate_clustering(labels, X):
    if len(set(labels)) > 1:  # Avoid cases where all points are in one cluster
        sil_score = silhouette_score(X, labels)
        db_score = davies_bouldin_score(X, labels)
        return sil_score, db_score
    else:
        return None, None

kmeans_labels = kmeans.fit_predict(X)
dbscan_labels = dbscan.fit_predict(X)

kmeans_sil, kmeans_db= evaluate_clustering(kmeans_labels, X)
dbscan_sil, dbscan_db = evaluate_clustering(dbscan_labels, X)

print("K-Means Clustering:")
print(f"Silhouette Score: {kmeans_sil:.4f}")
print(f"Davies-Bouldin Index: {kmeans_db:.4f}")

print("\nDBSCAN Clustering:")
print(f"Silhouette Score: {dbscan_sil:.4f}")
print(f"Davies-Bouldin Index: {dbscan_db:.4f}")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis')
ax1.set_title("K-Means Clustering")
ax2.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='plasma')
ax2.set_title("DBSCAN Clustering")

plt.show()
```
